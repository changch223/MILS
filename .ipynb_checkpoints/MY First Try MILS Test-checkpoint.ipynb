```python
import os
import torch
import numpy as np
from PIL import Image
from open_clip import create_model_and_transforms, get_tokenizer
from optimization_utils import (
    Scorer as S,
    Generator as G,
    get_text_features,
    get_image_features,
)
from paths import IMAGEC_COCO_IMAGES, OUTPUT_DIR
import transformers
import random
```

```python
# 參數設定
image_id = 42  # 可換成你想測試的圖片ID
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
import torch
print("Using device:", "cuda" if torch.cuda.is_available() else "cpu")
print("GPU available:", torch.cuda.is_available())
```

```python
clip_model_name = "ViT-SO400M-14-SigLIP"
pretrained = "webli"
text_model = "meta-llama/Meta-Llama-3.1-8B-Instruct"
init_descriptions_file = "init_descriptions/image_descriptions_per_class.txt"
prompt_file = "prompts/image_captioning_shorter.txt"
output_dir = os.path.join(OUTPUT_DIR, "notebook_test")
os.makedirs(output_dir, exist_ok=True)
```

```python
# 載入 clip 和 llm 模型
clip_model, _, preprocess = create_model_and_transforms(clip_model_name, pretrained=pretrained)
clip_model.to(device).eval()
tokenizer = get_tokenizer(clip_model_name)
```

```python
text_pipeline = transformers.pipeline(
    "text-generation",
    model=text_model,
    model_kwargs={"torch_dtype": torch.bfloat16},
    device_map=device,
)

# 安全設定 pad token
if text_pipeline.tokenizer.pad_token is None:
    text_pipeline.tokenizer.pad_token = text_pipeline.tokenizer.eos_token
```

```python
# 載入 prompt
with open(prompt_file, "r", encoding="utf-8") as f:
    text_prompt = f.read()

# 載入初始描述語料
with open(init_descriptions_file, "r", encoding="utf-8") as f:
    init_sentences = [line.strip() for line in f.readlines()]
init_sentences = random.sample(init_sentences, 50)
```

```python
IMAGEC_COCO_IMAGES = r"C:\Users\wayne\MILS\val2014"  # 改成你的實際路徑
# 圖片路徑
image_path = os.path.join(IMAGEC_COCO_IMAGES, f"COCO_val2014_{image_id:012}.jpg")
target_feature = get_image_features(clip_model, preprocess, [image_path], device, 1).detach().cpu().numpy()[0]
```

```python
# 定義 scorer function
def clip_scorer(sentences, target_feature):
    text_features = get_text_features(clip_model, tokenizer, sentences, device, 1, amp=True, use_format=False)
    return text_features.detach().cpu().numpy() @ target_feature
```

```python
scorers = {
    f"{image_id}": {
        "func": clip_scorer,
        "target_feature": target_feature
    }
}
scorer = S(scorers, batch_size=1, key=lambda x: -x, keep_previous=50)
```

```python
# 初始化初始分數
lines_with_scores = {
    f"{image_id}": [(s, l) for s, l in zip(scorer.score(str(image_id), init_sentences), init_sentences)]
}
```

```python
# 建立 Generator
generator = G(
    text_pipeline,
    text_model,
    requested_number=50,
    keep_previous=50,
    prompt=text_prompt,
    key=lambda x: -x[0],
    batch_size=1,
    device=device,
    exploration=0.0,
)
```

```python
# 最佳描述優化（少量迭代即可）
for i in range(3):  # 只跑3次就好
    new_lines = generator(lines_with_scores)
    lines_with_scores = scorer(new_lines)
```

```python
# 顯示最佳結果
best_score, best_caption = sorted(lines_with_scores[f"{image_id}"], key=lambda x: -x[0])[0]
print("Best Caption:", best_caption)
print("Score:", best_score)

# 顯示圖片
Image.open(image_path)
```
